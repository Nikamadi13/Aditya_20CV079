```python
def dupuit_discharge(K, b, h1, h2):
    """
    Calculate the discharge in a confined aquifer using Dupuit's equation.

    Parameters:
    - K: Hydraulic conductivity of the aquifer (in m/s).
    - b: Width of the aquifer (in meters).
    - h1: Hydraulic head at one end of the flow path (in meters).
    - h2: Hydraulic head at the other end of the flow path (in meters).

    Returns:
    - Q: Discharge (flow rate) in cubic meters per second (m³/s).
    """
    Q = -K * b * (h2 - h1)
    return Q

# Example usage:
K = 1e-2  # Hydraulic conductivity in m/s
b = 80  # Aquifer width in meters
h1 = 15  # Hydraulic head at the first point in meters
h2 = 7   # Hydraulic head at the second point in meters

discharge = dupuit_discharge(K, b, h1, h2)
print(f"Discharge: {discharge} m³/s")
```
