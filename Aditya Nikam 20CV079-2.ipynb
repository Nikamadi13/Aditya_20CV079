```python
def equipment_productivity(output_per_hour, working_hours_per_day, working_days):
    """
    Estimate the productivity of construction equipment.

    Parameters:
    - output_per_hour: Output (e.g., volume of earthwork) per hour (in cubic meters, tons, etc.).
    - working_hours_per_day: Number of hours the equipment works per day.
    - working_days: Number of working days in a month.

    Returns:
    - total_output: Total output in the given time frame (e.g., per month).
    """
    total_output = output_per_hour * working_hours_per_day * working_days
    return total_output

# Example usage:
output_per_hour = 20  # Example output per hour (cubic meters, tons, etc. per hour)
working_hours_per_day = 10  # Typical working hours per day
working_days = 25  # Number of working days in a month

estimated_output = equipment_productivity(output_per_hour, working_hours_per_day, working_days)
print(f"Estimated total output: {estimated_output} cubic meters (or tons, etc.)")
```
