```python
def trapezoidal_footing_volume(h, b1, b2, L):
    """
    Calculate the volume of concrete in a trapezoidal footing.

    Parameters:
    - h: Height (depth) of the trapezoidal footing (in meters).
    - b1: Length of the shorter base (in meters).
    - b2: Length of the longer base (in meters).
    - L: Length of the trapezoidal footing (in meters).

    Returns:
    - volume: Volume of concrete in the trapezoidal footing (in cubic meters).
    """
    volume = (h / 2) * (b1 + b2) * L
    return volume

# Example usage:
h = 2  # Height in meters
b1 = 2.0  # Shorter base length in meters
b2 = 4.0  # Longer base length in meters
L = 5.0  # Length of the trapezoidal footing in meters

concrete_volume = trapezoidal_footing_volume(h, b1, b2, L)
print(f"Total volume of concrete in the trapezoidal footing: {concrete_volume} cubic meters")

```
